```python
states = ['Andhra_Pradesh',
 'Arunachal_Pradesh',
 'Assam',
 'Bihar',
 'Chhattisgarh',
 'Goa',
 'Gujarat',
 'Haryana',
 'Himachal_Pradesh',
 'Jammu & Kashmir',
 'Jharkhand',
 'Karnataka',
 'Kerala',
 'Madhya_Pradesh',
 'Maharashtra',
 'Manipur',
 'Meghalaya',
 'Mizoram',
 'Nagaland',
 'Odisha',
 'Punjab',
 'Rajasthan',
 'Sikkim',
 'Tamil_Nadu',
 'Telangana',
 'Tripura',
 'Uttar_Pradesh',
 'Uttarakhand',
 'West_Bengal1',
 'Andaman & Nicobar Islands',
 'Chandigarh',
 'Delhi',
 'Puducherry',
 'All_India GDP']
```

```python
import os
```

```python
gsva=pd.DataFrame()
```

```python
path = ("C:/Data Science/Case Studies/3 GDP Analysis with Niti Ayog/Datasets/GSVA//")
```

```python
listofFiles=os.listdir(path)
listofFiles
```

```python
len(listofFiles)
```

```python
# Reading mutliple files and combining those files for one selected index value and tranposing into cloumn format
for filename in listofFiles:
    for statename in states:
        if statename in filename:
            filepath = path + filename
            data = pd.read_csv(filepath,encoding="ISO-8859-1")
            data = data[["Item","2014-15"]]
            data = data.T
            data.columns = data.iloc[0]
            data = data[1:]
            data["State"] = statename
            gsva = pd.concat([data,gsva])
```

```python
gsva.set_index("State",inplace=True)
```
